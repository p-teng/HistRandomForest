```python
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
import math
import timeit
import matplotlib.pyplot as plt
```

```python
data = load_iris()
X, y = make_classification(n_samples=500000, n_features=8,
                            n_informative=6, n_redundant=0,
                            random_state=0, shuffle=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
hrf_acc = []
hrf_time = []
x_bins = 0
num_bins = []
for i in range(8):
    x_bins += 1
    num_bins.append(x_bins)
    n_bins = 2 ** x_bins
    if (n_bins == 256):
        n_bins -= 1

    HRF = HistRandomForestClassifier(max_bins=n_bins, n_estimators=50, min_samples_leaf=20)
    start = timeit.default_timer()
    HRF.fit(X_train, y_train)
    time=timeit.default_timer() - start

    y_pred = HRF.predict(X_test)
    acc = accuracy_score(y_test, y_pred)

    hrf_time.append(time)
    hrf_acc.append(acc)

    print("# of Bins: " + str(n_bins) + "\tAcc: " + str(acc) + "\tTime: " + str(time))
```

```python
print("Bagged Decision Tree Classifier")
from sklearn.tree import DecisionTreeClassifier
DTclf = DecisionTreeClassifier(max_depth=250, min_samples_leaf=20)
start = timeit.default_timer()
rf = BaggingClassifier(base_estimator=DTclf, n_estimators=50, random_state=1).fit(X_train, y_train)
time=str(timeit.default_timer() - start)
print("Time: " + str(time))
y_pred = rf.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print("Acc: " + str(acc))
rf_acc = [float(acc)] * len(num_bins)
rf_time = [float(time)] * len(num_bins)
```

```python
print("Vanilla Random Forest Classifier")
start = timeit.default_timer()
RFclf = RandomForestClassifier(n_estimators=50)
RFclf.fit(X_train, y_train)
time=str(timeit.default_timer() - start)
print("Time: " + str(time))
y_pred = RFclf.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print("Acc: " + str(acc))
real_rf_acc = [float(acc)] * len(num_bins)
real_rf_time = [float(time)] * len(num_bins)
```
